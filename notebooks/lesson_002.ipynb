### Plotting Loss

In the previous lesson's exercise, we wrote the `average_absolute_error` function which essentially takes the predictions generated by our weights and the actual house prices, and tells us how "off" our predictions are. Based on the output of that function, we are then able to adjust our weights, try predicting again, and hopefully get a better result. This type of function is called a **loss function**. It tells us how much the **loss**, or gap, between our predictions and the true values is. We used one specific way to calculate loss, namely, the average of absolute differences across all training examples. There are many different ways to calculate loss depending on the application, but a loss function is a critical part of every machine learning model's training algorithm. It's how we determine if our current weights are "good" or not. The goal (and definition, really) of every training algorithm is to adjust our weights until we make our loss function output the lowest possible value.

How do we know by how much to adjust our weights, and in which direction? To answer this question, it would be helpful to visualize our loss function. In order to visualize our loss function easily, let's simplify our problem, and try to predict house sale prices based only on one feature, it's square footage. We'll slightly modify our loss function to reflect this change. We'll have a static list of square footage of different houses, a list of corresponding actual sale prices of the houses, and a static base_price. We'll also square each difference instead of taking it's absolute value. (This loss function is called **mean square error** or **MSE**, and has several benefits over the mean absolute error that we used last time. The benifits of this specific loss function are not important to know for now.) Then the function takes a value for the price per square foot, and outputs a loss value for that price.

```python
sqft = [800, 1200, 1500, 2000, 2500]
actual_prices = [150000, 210000, 260000, 340000, 420000]
base_price = 50000

def mean_square_error(price_per_sq_foot):
    predictions = [s * price_per_sq_foot for s in sqft]
    sum_of_errors = sum([(predictions[i] - actual_prices[i]) ** 2 for i in range(len(actual_prices))])
    return sum_of_errors / len(actual_prices)
for i in range(50):
  print([mean_square_error(i*10)],i*10)
```

We can then plot the input and output of this loss function to visualize it. We can see that the graph is a nice U shape. The lowers point is at around 170, which means that setting price per sq foot to about 170 will give us the lowest lost, and the best predictions.

<img src="https://github.com/amistein/ml-study-group/blob/main/images/loss_plot.png?raw=1" width="500">

We can also see that as the price per sqft increases or decreases from about 170 the loss goes up. The slope (steepness) of the curve shows us how quickly the loss will increase or decrease relative to the price per sqft. (Less quickly closer to the minimum loss, and more quickly further from it). Now, for any loss function and parameter value we have, we can plot the function and it's input values, and figure out in which direction to adjust our input so that we get to the minimum. We can also see whether should move by a little or a lot based on the slope of the curve. So in our example here, we know that if our price per sqft is 50, we need to move higher to get to our minimum.
